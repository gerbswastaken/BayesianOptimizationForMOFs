# random search

```python
import numpy as np
import pickle
import torch
import time
```

```python
# load data from `hydrogen_update_XY.ipynb`.
X = pickle.load(open('hydrogen_input_output.pkl', 'rb'))['x']
print("shape of X:", np.shape(X))

y = pickle.load(open('hydrogen_input_output.pkl', 'rb'))['y']
print("shape of y:", np.shape(y))

nb_COFs = pickle.load(open('hydrogen_input_output.pkl', 'rb'))['nb_COFs']
print("# COFs:", nb_COFs)

nb_runs = pickle.load(open('hydrogen_input_output.pkl', 'rb'))['nb_runs']
print("# runs:", nb_runs)

nb_iterations = pickle.load(open('hydrogen_input_output.pkl', 'rb'))['nb_iterations']
print("# iterations:", nb_iterations)
```

```python
# at each iteration, acquire a COF at random from the un-acquired set (hence, sample without replacement).
rs_res = dict()
rs_res['ids_acquired']     = []
for r in range(nb_runs):
    ids_acquired = np.random.choice(range(nb_COFs), replace=False, size=nb_iterations)
    rs_res['ids_acquired'].append(ids_acquired)
```

write results to file

```python
with open('rs_results_hydrogen.pkl', 'wb') as file:
    pickle.dump(rs_res, file)
```
